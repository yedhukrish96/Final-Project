```python
import requests
```

```python
import json
import pandas as pd
```

```python
import time
import random
from random import randint
from time import sleep

# Define the GraphQL query
graphql_query = '''
{
  lend {
    loans(limit:%d,offset:%d) {
      values {
        terms {
          currency
        }
        lenders {
          totalCount
        }
        id
        loanFundraisingInfo {
          fundedAmount
        }
        loanAmount
        activity {
          name
        }
        sector {
          name
        }
        use
        geocode {
          country {
            isoCode
            name
            region
          }
        }
        gender
        repaymentInterval
        disbursalDate
        fundraisingDate
        raisedDate
        lenderRepaymentTerm
      }
    }
  }
}



'''

# Define the base URL for the Kiva API
base_url = 'https://api.kivaws.org/graphql?query='

# Define the number of results to retrieve per request
batch_size = 100

# Define the total number of results to retrieve
total_results = 7000

# Calculate the number of requests to make
num_batches = int(total_results / batch_size)

# Define an empty list to store the loan data
loan_data = []

# Define a check number to keep track of how many loan records have been collected
check_num = 0

# Make requests to the API and store the loan data
for i in range(num_batches):
    # Calculate the offset for the current batch
    offset = i * batch_size
    
    # Combine the GraphQL query and the base URL
    combined_url = base_url + graphql_query % (batch_size, offset)
    
    # Send a request to the API with the combined URL
    response = requests.post(combined_url)
    
    # Parse the response and store the relevant data
    data = response.json()
    loans = data['data']['lend']['loans']['values']
    
    for loan in loans:
        use = loan['use']
        curr = loan['terms']['currency']
        sector = loan['sector']['name']
        loan_amount = loan['loanAmount']
        funded_amt = loan['loanFundraisingInfo']['fundedAmount']
        gender = loan['gender']
        repay = loan['repaymentInterval']
        region= loan['geocode']['country']['region']
        disburse = loan['disbursalDate']
        posted = loan['fundraisingDate']
        funded = loan['raisedDate']
        repay_time = loan['lenderRepaymentTerm']
        activity = loan['activity']['name']
        country_iso_code = loan['geocode']['country']['isoCode']
        country_name = loan['geocode']['country']['name']
        lender_count = loan['lenders']['totalCount']
        
       
    

        # Append the loan data to the list
        loan_data.append({
            'use': use,
            'currency': curr,
            'sector': sector,
            'loan_amount': loan_amount,
            'funded_amount': funded_amt,
            'gender': gender,
            'repayment_interval': repay,
            'region': region,
            'disbursed_time': disburse,
            'posted_time': posted,
            'funded_time': funded,
            'term_in_months': repay_time,
            'activity': activity,
            'country_iso_code': country_iso_code,
            'country_name': country_name,
            'lender_count': lender_count,
            
        })
        
        # Increment the check number by 1 for each loan record collected
        check_num += 1

        # Print the check number every 100 records
        if check_num % 100 == 0:
            print(f"{check_num} records collected")

    # Add a delay to avoid being blocked by the server
    sleep(randint(1,6))

# Convert the loan data to a dataframe
kiva_df = pd.DataFrame(loan_data)

# Print the first few rows of the dataframe
print(kiva_df.head())
```

```python
import pandas as pd
import numpy as np

def replace_nan_with_mode(df, col_name):
    # Replace the "nan" string with NaN values
    df[col_name] = df[col_name].replace('nan', np.nan)

    # Find the most common value using the mode method
    most_common = df[col_name].mode()[0]

    # Use fillna() method to replace NaN values with the most common value
    df[col_name] = df[col_name].fillna(most_common)

    return df
```

```python

def replace_month_values(df, columns):
    month_dict = {
        1: 'January',
        2: 'February',
        3: 'March',
        4: 'April',
        5: 'May',
        6: 'June',
        7: 'July',
        8: 'August',
        9: 'September',
        10: 'October',
        11: 'November',
        12: 'December'
    }
    for col in columns:
        df[col] = df[col].replace(month_dict)
    return df

```

## Here, I start with my EDA.

```python
import pandas as pd
import matplotlib.pyplot as plt


d1 = kaggle_df2.groupby('sector').agg(nr=('activity', 'count')).reset_index()
d1 = d1.sort_values('nr', ascending=False)

plt.figure(figsize=(8, 6))
plt.barh(d1['sector'], d1['nr'], color='sandybrown', edgecolor='black')
plt.gca().invert_yaxis()
plt.xlabel('Number of loans')
plt.title('Sector (all)')
plt.show()
```

 - These are the top 20 Sectors in terms of loans.
 - Agriculture is the sector with the most number of loans, followed by Food, Retail, Services etc.

```python

d2 = (
    kaggle_df2.groupby('activity')
            .agg(nr=('sector', 'count'))
            .nlargest(20, columns='nr')
            .reset_index()
            .assign(activity=lambda df: pd.Categorical(df['activity'], 
                                                        df['activity'][::-1],
                                                        ordered=True))
)

plt.figure(figsize=(8, 6))
plt.barh(d2['activity'], d2['nr'], color='yellowgreen', edgecolor='black')
plt.gca().invert_yaxis()
plt.xlabel('Number of loans')
plt.title('Activities (top 20 by number of loans)')
plt.show()
```

```python
d3 = (
    kaggle_df2.groupby('country')
            .agg(nr=('sector', 'count'))
            .nlargest(20, columns='nr')
            .reset_index()
            .assign(country=lambda df: pd.Categorical(df['country'], 
                                                       df['country'][::-1],
                                                       ordered=True))
)

plt.figure(figsize=(8, 6))
plt.barh(d3['country'], d3['nr'], color='tomato', edgecolor='black')
plt.gca().invert_yaxis()
plt.xlabel('Number of loans')
plt.title('Countries (top 20 by number of loans)')
plt.show()
```

```python
d4 = (
    kaggle_df2.groupby('region')
            .agg(nr=('sector', 'count'))
            .nlargest(20, columns='nr')
            .reset_index()
            .assign(country=lambda df: pd.Categorical(df['region'], 
                                                       df['region'][::-1],
                                                       ordered=True))
)

plt.figure(figsize=(8, 6))
plt.barh(d4['region'], d4['nr'], color='olive', edgecolor='black')
plt.gca().invert_yaxis()
plt.xlabel('Number of loans')
plt.title('Regions (top 20 by number of loans)')
plt.show()
```
